## Data Cleaning

Our first step is to clean the data and prepare it for analysis.

The following functions are used:

1. <ins>Clean_columns function<ins>
    * Column names are striped of extra whitespace and special characters
    * Underscores ('_') are used to join names together
    * Column names are converted to title case
    * Specific columns are specially renamed
    
    <break></break>
2. <ins>Convert_and_sort function<ins>
    * The 'Time' column is converted to a datetime dtype
    * The Dataframe is sorted by the 'Time' column in ascending order
    * The Dataframe has the index reset

```python
# Adjust column names to be uniform
def clean_column_names(df):
    # Rename columns by removing special characters using regex
    df.columns = exchange_rates.columns.str.replace(
        pat=r"[(^\[)(\]$)]", repl="", regex=True
    ).str.strip()

    # Replace spaces with underscores and title case the column names
    df.columns = df.columns.str.replace(pat=" ", repl="_").str.title()

    # Rename specific columns
    df.rename(columns={"Us_Dollar": "US_Dollar", "Period\\Unit:": "Time"}, inplace=True)
```

```python
# Converting a column to datetime and sorting the DataFrame by the supplied column
def convert_and_sort(df, col):
    # Convert supplied column to datetime
    df[col] = pd.to_datetime(df[col])

    # Sort DataFrame by supplied column
    df.sort_values(col, inplace=True)

    # Reset index after sorting
    df.reset_index(drop=True, inplace=True)
```

```python
# Make all column names the same
clean_column_names(df=exchange_rates)

# Convert the "Time" column to datetime and sort the DataFrame
convert_and_sort(df=exchange_rates, col="Time")
```

## Data Processing and Exploration

The data needs to processed and explored to gain any insights on the Euro-Dollar Exchange Rate during specific periods. 

The following process will be followed:

1. A subset of the original dataset will be extracted.
2. All irrelevant/missing values will be removed.
3. Column dtypes will be modified as required.
4. Any addtional column(s) will be added as required.
5. Queries will be performed based on insights required.

```python
# Extracting the "Time" and "US_Dollar" columns from the exchange_rates DataFrame
euro_to_dollar = exchange_rates[["Time", "US_Dollar"]].copy()
```

```python
# Find the row indexs where the value in the "US_Dollar" column is '-'
rows_to_drop = euro_to_dollar.query(expr="US_Dollar == '-'", inplace=False).index

# Removing the rows with the found indices
euro_to_dollar.drop(labels=rows_to_drop, axis=0, inplace=True)
```

```python
# Convert the "US_Dollar" column to numeric values using 'pd.to_numeric'
euro_to_dollar["US_Dollar"] = pd.to_numeric(euro_to_dollar["US_Dollar"])
```

### Rolling Mean

After cleaning the Data, plotting a rolling mean (or 'moving average') of the euro-dollar exchange rate shows the daily variation that occurs.

```python
# Calculate the rolling mean of the US Dollar exchange rate over a window of 30 days
euro_to_dollar["Rolling_Mean"] = euro_to_dollar["US_Dollar"].rolling(window=30).mean()

# Rounding the rolling mean to 2 decimal places for querying
euro_to_dollar["Rolling_Mean_Rounded"] = round(
    number=euro_to_dollar["Rolling_Mean"], ndigits=2
)
```

```python
plt.plot("Time", "Rolling_Mean", data=euro_to_dollar)
plt.show()
```

## Data Visualisation

### 2007 - 2008 Financial Crisis

The fluctuations of the euro-dollar exchange rate during the 2007 - 2008 financial crisis can be shown against the euro-dollar exchange rate from 2006 - 2009.

```python
# Filter the euro_to_dollar DataFrame for the years 2007 and 2008
financial_crisis_07_08 = euro_to_dollar.query(
    expr="2007 <= Time.dt.year <= 2008", inplace=False
).reset_index(drop=True)

# Filter the euro_to_dollar DataFrame for the years 2006 to 2009
financial_crisis = euro_to_dollar.query(
    expr="2006 <= Time.dt.year <= 2009", inplace=False
).reset_index(drop=True)
```

## Identifying the Crisis Period

```python
# Finding worst period of financial crisis visually
financial_crisis_07_08.plot(x="Time", y="Rolling_Mean")
plt.grid(visible=True, alpha=0.5)
plt.show()
```

Plotting the rolling mean for the 2007 - 2008 financial crisis gives an indication that March to September of 2008 was the worst period for the euro-dollar exchange rate. The rolling means were approximatley 1.48 and 1.45 respectfully.

```python
# Confirming worst period by checking frequency of exchange rates for year and month
financial_crisis_07_08.groupby(
    by=[
        financial_crisis_07_08["Time"].dt.strftime(date_format="%y-%m"),
        "Rolling_Mean_Rounded",
    ]
).size()
```

```python
# Querying dataframe for indexes of dates for crisis peak and end
display('crisis_peak_start',financial_crisis_07_08.query("Time.dt.strftime('%y-%m')=='08-03'").head(1))
display('crisis_peak_end',financial_crisis_07_08.query("Time.dt.strftime('%y-%m')=='08-09'").tail(1))
```

```python
# Using indexes to retreive and store dates for crisis peak and end
crisis_peak_start = financial_crisis_07_08.iloc[298]["Time"]
crisis_peak_end = financial_crisis_07_08.iloc[442]["Time"]
```

## Plotting the Financial Crisis

```python
# Create subplots
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 3))

# Plot the Rolling_Mean data for financial_crisis
ax.plot("Time", "Rolling_Mean", data=financial_crisis, linewidth=1, color="blue")

# Plot the Rolling_Mean data for financial_crisis_07_08
ax.plot("Time", "Rolling_Mean", data=financial_crisis_07_08, linewidth=3, color="red")

# Get the x-axis position for each xtick label and store it in a list
xtick_positions = [x.get_position()[0] for x in ax.get_xmajorticklabels()]

# Set the xticks and xticklabels at correct positions
ax.set(
    xticks=xtick_positions,
    xticklabels=["2006", "", "2007", "", "2008", "", "2009", "", "2010"],
)

# Disable bottom ticks and adjust the transparency of the grid
ax.tick_params(bottom=False, grid_alpha=0.5)

# Add a title
plt.suptitle(
    t="Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis",
    x=0.49,
    va="baseline",
    color="black",
    weight="bold",
    fontsize=13,
)

# Add a subtitle
plt.title(
    label="Euro-USD exchange rates between 2006 and 2010",
    loc="left",
    color="black",
    fontsize=12,
)

# Add a shaded region to highlight the crisis peak
# 'date2num' method used to convert date to x-axis position automatically
ax.axvspan(
    xmin=date2num(crisis_peak_start),
    xmax=date2num(crisis_peak_end),
    alpha=0.2,
    color="grey",
)

# Add a signature
ax.text(
    x=13093,
    y=1.08,
    s="Akhil Rambaran" + " " * 81 + "Source: European Central Bank",
    fontsize=10,
    color="white",
    backgroundcolor="grey",
)

# Show the plot
plt.show()
```
